# Boxplot of Magnitude column

A boxplot is a graph that gives a visual indication of how a data set's 25th percentile, 50th percentile, 75th percentile, minimum, maximum and outlier values are spread out and compare to each other.

```python
sns.boxplot(x=s['Year'],palette='hls')
sns.despine()
plt.show()
```

```python
sns.boxplot(x=s['Selling_price'],palette='hls')
sns.despine()
plt.show()
```

# To remove outliars

```python
q1=s["Selling_price"].quantile(0.25)
q3=s["Selling_price"].quantile(0.75)
iqr=q3-q1
lower_limit = q1 - 3.5*iqr

upper_limit = q3 + 0.5*iqr

lower_limitoutliers = s[s['Selling_price'] < lower_limit]

upper_limitoutliers = s[s['Selling_price'] > upper_limit]
new_df= s[(s['Selling_price'] < upper_limit)&(s['Selling_price'] >lower_limit)]

import matplotlib.pyplot as plt

import seaborn as sns

sns.set(rc={"figure.figsize": (6, 6)})

sns.boxplot(x=new_df["Selling_price"])

plt.legend(loc="best")
```

# Dist Plot of Year column
